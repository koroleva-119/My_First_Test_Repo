```python
a=int(input('Введите целое число от 0 до 5 \n'))
if a==0:
    print('Ноль')
elif a==1:
    print('Один')
elif a==2:
    print('Два')
elif a==3:
    print('Три')
elif a==4:
    print('Четыре')
else:
    print('Некорректный ввод')
```
