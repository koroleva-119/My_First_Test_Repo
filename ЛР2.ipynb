```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url = "https://drive.google.com/uc?export=download&id=1J_9c0QafYo3XC6X62OxQLVUpeIv2qCVy"
input_df = pd.read_csv(url) 
print('Просмотр первых строк таблицы:')
input_df.head() 
```

```python
print('Просмотр последних строк таблицы:')
input_df.tail()
```

```python
print('Временной ряд выработки ВЭС:')
wind_power = input_df.iloc[: , 0].values

from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

output_notebook()

p = figure(plot_width = 800, plot_height = 400)

p.line(np.arange(len(wind_power)), wind_power, line_width = 1, color='green')
p.scatter(np.arange(len(wind_power)), wind_power, color='green')
p.line(np.arange(len(wind_power)), line_width = 1, color = 'green')
p.scatter(np.arange(len(wind_power)), color = 'green')

output_file("test.html")

show(p)
```

```python
print('Временной ряд выработки СЭС:')
solar_power = input_df.iloc[: , 1].values

output_notebook()

p = figure(plot_width = 800, plot_height = 400)

p.line(np.arange(len(solar_power)), solar_power, line_width = 1)
p.scatter(np.arange(len(solar_power)), solar_power)
p.line(np.arange(len(solar_power)), line_width = 1)
p.scatter(np.arange(len(solar_power)))

output_file("test.html")

show(p)
```

```python
print('Временной ряд потребления:')
consumption = input_df.iloc[: , 2].values

output_notebook()

p = figure(plot_width = 800, plot_height = 400)

p.line(np.arange(len(consumption)), consumption, line_width = 1, color = 'orange')
p.scatter(np.arange(len(consumption)), consumption, color = 'orange')
p.line(np.arange(len(consumption)), line_width = 1)
p.scatter(np.arange(len(consumption)))

output_file("test.html")

show(p)
```

```python
print('График разницы потребления и выработки активного потребителя')
y=input_df['Wind_power_MW']+input_df['Solar_power_kW']-input_df['Consumption_MW']

output_notebook()

p = figure(plot_width = 800, plot_height = 400)

p.line(np.arange(len(y)), y, line_width = 1, color = 'red')
p.scatter(np.arange(len(y)), y, color = 'red')
p.line(np.arange(len(y)), line_width = 1)
p.scatter(np.arange(len(y)))

output_file("test.html")

show(p)
```

```python
# какое количество электроэнергии потребитель сможет продать во внешнюю систему
print(f'Потребитель сможет продать {y.sum()} МВ*ч во внешнюю систему')
```
