```python
import math
import sqlite3

def dist():
  # Ввод координат двух точек
  x1, y1, x2, y2 = map(float, input("Введите координаты двух точек (x1 y1 x2 y2) через пробел: ").split())

  # Расстояние между точками
  distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
  conn = sqlite3.connect("example.db")
  cursor = conn.cursor()
  cursor.execute("CREATE TABLE IF NOT EXISTS distances (id INTEGER PRIMARY KEY AUTOINCREMENT, distance REAL)")
  cursor.execute("INSERT INTO distances (distance) VALUES (?)", (distance,))
  cursor.execute('SELECT distance FROM distances')
  row = cursor.fetchone()
  print(f"Расстояние между точками: {row[0]:.5f}")
  conn.commit()
  conn.close()

def min_max():
  # Ввод чисел a, b, c, d
  a, b, c, d = map(float, input("Введите числа a, b, c, d через пробел: ").split())

  # Вычисление минимального и максимального значений
  min_value = min(a, b, c, d)
  max_value = max(a, b, c, d)

  # Вывод результата
  print(f"Минимальное значение: {min_value}")
  print(f"Максимальное значение: {max_value}")

  # Сохранение результата в базу данных SQLite
  conn = sqlite3.connect("example.db")
  cursor = conn.cursor()
  cursor.execute("CREATE TABLE IF NOT EXISTS min_max (id INTEGER PRIMARY KEY AUTOINCREMENT, min_value REAL, max_value REAL)")
  cursor.execute("INSERT INTO min_max (min_value, max_value) VALUES (?, ?)", (min_value, max_value))
  conn.commit()
  conn.close()

def set_dict():
  # Ввод словаря
  key = input('Введите ключи: ').split()
  value = input('Введите значения: ').split()
  my_dict = dict(zip(key, value))

  # Таблица для хранения ключей словаря
  conn = sqlite3.connect("example.db")
  cursor = conn.cursor()
  cursor.execute('''CREATE TABLE IF NOT EXISTS dictionary
                    (id INTEGER PRIMARY KEY AUTOINCREMENT,
                    key TEXT NOT NULL)''')
  for key in my_dict.keys():
      cursor.execute("INSERT INTO dictionary (key) VALUES (?)", (key))
  conn.commit()

  # Вывести ключи словаря
  cursor.execute("SELECT key FROM dictionary")
  rows = cursor.fetchall()
  print(rows)

  # Множества
  my_set = input('Введите элементы множества: ').split()
  print('Добавьте в множество 3 новых элемента')
  for i in range (3):
    l = input("Введите новый элемент: ")
    my_set.append(l)
  my_set.remove(input('Удалите существующий элемент множества: '))

  # Таблица для хранения множества
  cursor.execute('''CREATE TABLE IF NOT EXISTS myset
                    (element text)''')
  for element in my_set:
      cursor.execute("INSERT INTO myset (element) VALUES (?)", (element))
  conn.commit()
  cursor.execute("SELECT element FROM myset")
  rows = cursor.fetchall()
  print(rows)
  conn.close()

while True:
  print('Выберите действие:')
  print('1. Вычисление расстояния между двумя точками')
  print('2. Работа с множеством и словарем')
  print('3. Определение минимального и максимального значения')
  print('4. Выход')
  choice = input()
  if choice == '1':
    dist()
  elif choice == '2':
    set_dict()
  elif choice == '3':
     min_max()
  elif choice == '4':
    break
  else:
    print('Некорректный выбор')
```
